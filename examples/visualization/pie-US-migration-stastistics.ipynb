## Percentage of permanent residence issued to Chinese and Indian nationals in 2015, comparing with the rest of Asia

Data source: [Yearbook of Immigration Statistics 2015, Homeland Security, United States](https://www.dhs.gov/immigration-statistics/yearbook/2015)

```python
from pyecharts import online
online()

import pyexcel as p

sheet=p.get_sheet(file_name="fy2015_table3d.xls")
sheet.top()
```

Let skip the first three rows because they contain meta data only.

```python
sheet=p.get_sheet(file_name="fy2015_table3d.xls", start_row=3)
sheet.top()
```

```python
sheet.name_rows_by_column(0)
sheet.row.select(['Region and country of birth','Asia', 'India', 'China, People\'s Republic'])
sheet
```

```python
sheet
```

```python
sheet.plot.line()
```

```python
sheet.transpose()
sheet
```

```python
pie_dict = {}
pie_dict['China'] = float(sheet['2015', 'China, People\'s Republic'])/sheet['2015', 'Asia']
pie_dict['India'] = float(sheet['2015', 'India'])/sheet['2015', 'Asia']
pie_dict['The Rest of Asia'] = 1 - pie_dict['China'] - pie_dict['India']
pie_sheet = p.get_sheet(adict=pie_dict)
pie_sheet
```

```python
pie_sheet.title = 'Percentage of permanent residence issued to Chinese and Indian nationals in 2015, comparing with the rest of Asia'

pie_sheet.plot.pie()
```
