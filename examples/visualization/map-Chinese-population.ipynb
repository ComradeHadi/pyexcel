# Chinese Census 2010

Data source
http://www.stats.gov.cn/tjsj/pcsj/rkpc/6rp/excel/A0101a.xls

```python
from pyecharts import online
online()


import pyexcel as p

sheet = p.get_sheet(file_name="A0101a.xls", start_row=8, row_limit=31, column_limit=2)
```

```python
sheet.column.format(0, lambda name: name.replace(' ', ''))

def convert_to_int(n):
    if not isinstance(n, int):
        n =n.replace(u'，', '')
        return int(n)
    return n
        
sheet.column.format(1, convert_to_int)
sheet
```

```python
sheet.name = "Population in 31 provinces and cities"
sheet.plot.map(maptype='china',
               is_visualmap=True, 
               visual_range=[min(sheet.column[1]), max(sheet.column[1])], 
               is_map_symbol_show=False)
```
